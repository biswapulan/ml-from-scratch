```python
import numpy as np
import matplotlib.pyplot as plt

#Dataset
X = np.array([1,2,3,4,5], dtype=float)
y = np.array([2,4,6,8,10], dtype=float)

#Initialize
w = 0.0
b = 0.0
lr = 0.001
epochs = 1000
# Try changing this epochs and you get w approches 2.0

# Full code for epochs
for epoch in range(epochs):
  #Step 1 - predict
  y_hat = w * X + b

  #Step 2 - Loss
  mse = np.mean((y - y_hat) ** 2)

  #Step 3 - Gradients
  dw = (-2/len(X)) * np.sum(X * (y - y_hat))
  db = (-2/len(X)) * np.sum(y - y_hat)

  #Step 4 - Update
  w = w - (lr * dw)
  b = b - (lr * db)

  if epoch % 100 == 0:
    print(f"Epoch {epoch} | Loss: {mse:.4f} | w: {w:.4f} | b: {b:.4f}")
```

```python
#plot actual vs predicte
plt.scatter(X,y, color='blue', label='Actual')
plt.plot(X, w*X + b, color="red", label='Predicted')
plt.legend()
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```
