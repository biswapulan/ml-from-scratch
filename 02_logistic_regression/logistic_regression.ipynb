```python
import numpy as np
#Dataset - is tumor malignant ? (1) or not (0)
X = np.array([1,2,3,4,5], dtype=float)
y = np.array([0,0,0,1,1], dtype=float)

#Initialize
w = 0.0
b = 0.0
lr = 0.01
epochs = 1000

#1.Implement sigmoid function
def sigmoid(z):
  return 1 / ( 1 + np.exp(-z))

for epoch in range(epochs):

  #2. Predict y_hat using sigmoid
  y_hat = sigmoid(w*X + b)

  #3. Compute binary cross entropy loss
  L = -(np.mean((y*np.log(y_hat)) + ((1-y)*np.log(1-y_hat))))


  #4. Compute gradients dw, db
  dw = np.mean(X * (y_hat - y))
  db = np.mean(y_hat - y)

  #5. Update w and b
  w = w - (lr*dw)
  b = b - (lr*db)

  #6. Print loss every 100 epochs
  if epoch % 100 == 0:
        print(f"Epoch {epoch} | Loss: {L:.4f} | w: {w:.4f} | b: {b:.4f}")
```

```python
print("\n--- Predictions ---")
for i in range(len(X)):
    prob = sigmoid(w*X[i] + b)
    pred = 1 if prob >= 0.5 else 0
    print(f"X={X[i]} | Probability: {prob:.4f} | Predicted: {pred} | Actual: {int(y[i])}")
```

```python
predictions = [1 if sigmoid(w*X[i] + b) >= 0.5 else 0 for i in range(len(X))]
accuracy = np.mean(np.array(predictions) == y) * 100
print(f"\nAccuracy: {accuracy:.2f}%")
```
