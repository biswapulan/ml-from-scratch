```python
import numpy as np
from collections import Counter
```

```python
#Dataset
X_train = np.array([[1,2],[2,3],[3,4],[6,7],[7,8],[8,9]], dtype=float)
y_train = np.array([0,0,0,1,1,1])

#New point to classify
X_test = np.array([5,6], dtype=float)

K = 3

#1. Compute euclidian distance from X_test to every point in X_train
def euclidian_distance(x1,x2):
  return np.sqrt(np.sum((x1 - x2) ** 2))

#2. Sort distances and pick K smallest indices
distances = [euclidian_distance(X_test, x) for x in X_train]
# for x in X_train:
#   distances.append(euclidian_distance(X_test, x))

#3. Sort distances and pick K smallest indices
k_indices = np.argsort(distances)

#4. Get labels of K nearest neighbors
labels = [y_train[k_indices[x]] for x in range(K)]
# for x in range(K):
#   labels.append(y_train[k_indices[x]])

#5. Vote - whichever class appears most is the prediction
most_common = Counter(labels).most_common(1)[0][0]

print(f"Test data {X_test}, belongs to class {most_common}")
```
